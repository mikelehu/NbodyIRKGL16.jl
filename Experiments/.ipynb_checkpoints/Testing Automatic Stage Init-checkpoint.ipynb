## Initial value problem: 5-body problem (outer solar system)

We consider $N=5$ bodies of the outer solar system: the Sun, Jupiter, Saturn, Uranus, and Neptune.
The initial values $u_{00}$ are taken from DE430, Julian day (TDB) 2440400.5 (June 28, 1969). 

```julia
PATH_ODES="../ODEProblems/"

include(string(PATH_ODES,"Initial5Body.jl"))
include(string(PATH_ODES,"Nbody.jl"))
include(string(PATH_ODES,"Nbody2nd.jl"));
```

```julia
u0, Gm, bodylist = Initial5Body(Float64)
u0_B, Gm_B, bodylist = Initial5Body(BigFloat)

q0=u0[:,:,1]
v0=u0[:,:,2]

N = length(Gm)

show(bodylist)
E0=NbodyEnergy(u0,Gm)
```

```julia
t0=0.
tF=3.e5 #3e4

tspan= (t0,tF)
tspan_B=(BigFloat(t0),BigFloat(tF))

udim=u0[1:1,1:1,1:1]
prob = ODEProblem(NbodyODE!, u0,tspan , Gm)
prob2nd = SecondOrderODEProblem(NbodyODE2nd!,v0,q0,(t0,tF),Gm)
prob_B = ODEProblem(NbodyODE!, u0_B, tspan_B, Gm_B);
```

### Integrations

```julia
tols=abstols=reltols=[1e-8, 1e-9, 1e-10, 1e-11, 1e-12, 1e-13, 1e-14, 1e-15]

dts4=[50., 75., 100., 125., 150., 175., 200., 225., 250.]
dts4p=dts4

dts8= [350., 400., 450., 500, 550, 600, 650, 700, 750.]
dts8p=dts8;
```

#### s=4,  partitioned=false

```julia
nruns=100


s=4
wp1=launch_IRKGL_simd_tests(final_state, prob, udim, s, dts4,
                            initial_interp=0, partitioned=false, nruns=nruns)
wp2=launch_IRKGL_simd_tests(final_state, prob, udim, s, dts4,
                            initial_interp=1, partitioned=false, nruns=nruns)
wp3=launch_IRKGL_simd_tests(final_state, prob, udim, s, dts8p,
                            initial_interp=2, partitioned=false, nruns=nruns)
wp4=launch_IRKGL_simd_tests(final_state, prob, udim, s, dts4p,
                            initial_interp=-1, partitioned=false, nruns=nruns)
```

```julia
all_times=vcat(wp1.times, wp2.times, wp3.times, wp4.times)

ydt=std(wp1.times)/10
yrange=(log10(minimum(all_times)-ydt), log10(maximum(all_times)+ydt))


plot(title="My benchmarks (s=4, partitioned=false)",xlabel="Error", ylabel="Time (s)", ylims=(10^(-3.1),10^(-1.8)))
# 
plot!(wp1.errors, wp1.times, seriestype=:scatter, scale=:log10, label="Init_Interp=0",color="red")
plot!(wp1.errors, wp1.times, scale=:log10, label="", lw=2, color="red")
#
k=1
plot!(wp2.errors, wp2.times, seriestype=:scatter, scale=:log10, label="Init_Interp=1",color="green")
plot!(wp2.errors, wp2.times, scale=:log10, label="", lw=2, color="green")
#
plot!(wp3.errors, wp3.times, seriestype=:scatter, scale=:log10, label="Init_Interp=2",color="orange")
plot!(wp3.errors, wp3.times, scale=:log10, label="", lw=2, color="orange")
#
plot!(wp4.errors, wp4.times, seriestype=:scatter, scale=:log10, label="Init_Interp=-1",color="blue")
plot!(wp4.errors, wp4.times, scale=:log10, label="", lw=2, color="blue")
#
```
