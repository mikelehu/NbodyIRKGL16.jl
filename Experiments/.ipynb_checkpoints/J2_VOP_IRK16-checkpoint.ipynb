## Initial value problem: VOP formulation of $J_2$-KS equation



\begin{equation}
\begin{split}
\frac{d}{d\tau} q &= r \, v, \quad q(0) = q_0,\\
\frac{d}{d\tau} v &= -\frac{\mu}{r^2} q -  r\, \nabla V(q), \quad v(0)=v_0
\end{split}
\tag{1}
\end{equation}

where 

\begin{equation}
q_0 = \left(
\begin{matrix}
x_0\\
y_0\\
z_0
\end{matrix}
\right), \quad 
v_0 = \left(
\begin{matrix}
\dot x_0\\
\dot y_0\\
\dot z_0
\end{matrix}
\right), \quad r=\|q\|, 
\end{equation}

and

\begin{equation}
V(q) =  \frac{C}{2\, r^3} \,\left(3\, \sin^2\theta -1 \right), \quad \sin\theta = \frac{z}{r}, \quad C=J_2\, \mu\, R_e^2.
\end{equation}

We have that
\begin{equation*}
\nabla V(x,y,z) = \frac{3\, C}{2\, r^5} 
\left(
\begin{matrix}
(1 - 5\, \sin^2\theta)\, x\\
(1 - 5\, \sin^2\theta)\, y \\
(3 - 5\, \sin^2\theta)\, z
\end{matrix}
\right).
\end{equation*}

The energy $\frac12 \, \langle v, v \rangle - \frac{\mu}{r} + V(q)$ is an invariant of the equations of motion. 
In addition, the $z$-component of the angular momentum
\begin{equation*}
x\, \dot y - \dot x\, y
\end{equation*}
is also a first integral.

We will rewrite the equations by making use of Kustaanheimo-Stiefel (KS) transformation: the new variables $u=(u_1, u_2, u_3,u_4) \in \mathbb{R}^4$ are related to the position variables $q = (x,y,z) \in \mathbb{R}$ by
\begin{equation*}
q = L(u) u,
\end{equation*}
where 

\begin{equation*}
L(u) = 
\left(
\begin{matrix}
u_1 & -u_2 & -u_3 & u_4 \\
u_2 & u_1 & -u_4 & -u_3 \\
u_3 & u_4 & u_1 & u_2 \\
u_4 & -u_3 & u_2 & -u_1
\end{matrix}
\right).
\end{equation*}
It holds that $\|u\|^2 = r$.

```julia
function V(q)
#   C = μ * J_2 * R_e^2   
    C = 1.7554962315534863e10
    z = q[3]
    r = sqrt(q[1]^2 + q[2]^2 + q[3]^2)    
    sinth = z/r
    return C*(3*sinth^2-1)/(2*r^3)
end

function gradV(q)
#   C = μ * J_2 * R_e^2   
    C = 1.7554962315534863e10
    x = q[1]
    y = q[2]
    z = q[3]
    r2 = x^2 + y^2 + z^2
    r = sqrt(r2)
    sinth = z/r
    aux1 = 1.5*C/(r*r2^2)
    aux2 = 5*sinth^2
    aux3 = aux1*(1 - aux2)
    return [aux3*x, aux3*y, aux1*(3-aux2)*z]
end
```

```julia
L(u) = [u[1] -u[2] -u[3]  u[4]
     u[2]  u[1] -u[4] -u[3]
     u[3]  u[4]  u[1]  u[2]
     u[4] -u[3]  u[2] -u[1] ]
```

Following~\cite{Stiefel-Scheifele}, the solution curve $q(\tau)$ of the initial value problem (1) can be parametrized as $q = L(u(\tau)) u(\tau)$,  where $(u(\tau),w(\tau))$ is the solution of the initial value problem

\begin{equation}
\begin{split}
\frac{d}{d\tau} u &= w, \quad u(0)=u_0,\\
\frac{d}{d\tau} w &= -\frac{h}{2} u -  \nabla R(u), \quad w(0)=w_0,
\end{split}
\tag{2}
\end{equation}


where

- $\displaystyle h= \frac{\mu}{\|q_0\|} - \frac12\, \langle \dot q_0, \dot q_0 \rangle -  V(q_0)$,


- $R(u) = \frac14\,  \|u\|^2\, V(L(u) u)$, that is,

\begin{equation*}
R(u) = \frac{1}{8\, r^2} \,\left(3\, \sin^2\theta -1 \right),
\end{equation*}

where  $r = u_1^2+u_2^2+u_3^2+u_4^2$ and
$\sin\theta = 2\,(u_1 u_3 + u_2 u_4)/r$,

```julia
function R(u)
#   C = μ * J_2 * R_e^2   
    C = 1.7554962315534863e10
    z = 2*(u[1]*u[3] + u[2]*u[4])
    r = u[1]^2 + u[2]^2 + u[3]^2 + u[4]^2
    sinth = z/r
    return C*(3*sinth^2-1)/(8*r^2)
end


function gradR(u)
#   C = μ * J_2 * R_e^2   
    C = 1.7554962315534863e10
    z = 2*(u[1]*u[3] + u[2]*u[4])
    r = u[1]^2 + u[2]^2 + u[3]^2 + u[4]^2
    w = 1/r^3
    sinth = z/r
    A = 0.5*C*w*(1 - 6*sinth^2)
    B = 1.5*C*w*sinth
    return [A*u[1]+B*u[3],A*u[2]+B*u[4],A*u[3]+B*u[1],A*u[4]+B*u[2]]
end
```

#### Determination of initial values

- $u_0 \in \mathbb{R}^4$ is chosen in such a way that 
\begin{equation*}
\left(
\begin{matrix}
x_0\\
y_0\\
z_0\\
0
\end{matrix}
\right)
 = L(u_0) u_0, 
\end{equation*}

- $w_0\in \mathbb{R}^4$ is determined as 
\begin{equation*}
w_0 = \frac12\, L(u_0)^T 
\left(
\begin{matrix}
\dot x_0\\
\dot y_0\\
\dot z_0\\
 0
 \end{matrix}
\right).
\end{equation*}


There are infinitely many choices of $u_0 \in \mathbb{R}^4$ for given $q_0 = (x_0,y_0,z_0)$. Among them, we will uniquely determine $u_0 =(u_{1,0}, u_{2,0}, u_{3,0}, u_{4,0})$ by imposing the following conditions:

- if $x_0\geq 0$, 
\begin{equation*}
u_{1,0} = u_{4,0}, \quad u_{1,0}, u_{4,0} \geq 0,
\end{equation*}
- if $x_0<0$, 
\begin{equation*}
u_{2,0} = u_{3,0}, \quad u_{2,0}, u_{3,0} \geq 0.
\end{equation*}



This uniquely determines a map $\chi:\mathbb{R}^3 \to \mathbb{R}^4$ such that $u_0 = \chi(q_0)$.
 More precisely, given $q = (x,y,z)$ and $r = \sqrt{x^2+y^2+z^2}$,  
 $u=(u_1,u_2,u_3,u_4) = \chi(q)$ is determined as follows:

- if $x\geq 0$, then
\begin{equation*}
u_1 = u_4 = \frac12\, \sqrt{r + x}, \quad 
u_2 = \frac{y u_1 + z u_4}{r + x}, \quad u_3 = \frac{z u_1 - y u_4}{r + x}, 
\end{equation*}
- if $x<0$, then 
\begin{equation*}
u_2 = u_3 = \frac12\, \sqrt{r - x}, \quad 
u_1 = \frac{y u_2 + z u_3}{r - x}, \quad u_4 = \frac{z u_2 - y u_3}{r - x}.
\end{equation*}

```julia
function χ(q)
    x = q[1]
    y = q[2]
    z = q[3]
    r = sqrt(x^2+y^2+z^2)
    if x >= 0
        aux = r + x
        u1 = 0.5*sqrt(aux)
        u4 = u1
        u2 = (y*u1 + z*u4)/aux
        u3 = (z*u1 - y*u4)/aux
    else
        aux = r - x
        u2 = 0.5*sqrt(aux)
        u3 = u2
        u1 = (y*u2 + z*u3)/aux
        u4 = (z*u2 - y*u3)/aux
    end
    return [u1, u2, u3, u4]
end
```

```julia
Ifcn(u,w) = u[4]*w[1]-u[3]*w[2]+u[2]*w[3]-u[1]*w[4]
Ifcn(U) = Ifcn(U[1:4],U[5:8])
```

```julia
q0 = [0., 37947.73745727695, 0.]
v0 = [3.297676220718193,0., 0.8244190551795483]
μ = 398600.8
h = μ/norm(q0) - 0.5*dot(v0,v0) -  V(q0) 
ω = sqrt(h/2)

u0 = χ(q0)
V0 = vcat(v0,[0.])
w0 = 0.5*L(u0)' * V0
r0 = norm(q0)

dim=length(size(u0))


# Konprobazioa:
(norm(L(u0)*u0 - vcat(q0,[0.])), norm(r0*V0 - 2* L(u0) * w0))
```

### VOP formulazioa



From now on, we assume that $h>0$, and denote $\omega=\sqrt{h/2}$.  The solution $(u(\tau), w(\tau))$ of the initial value problem (2) can be written as

\begin{align*}
u(\tau) &= \cos(\omega \tau) \alpha(\tau)+ \omega^{-1} \sin(\omega \tau) \beta(\tau),\\
w(\tau) &= -\omega\, \sin(\omega \tau) \alpha(\tau) + \cos(\omega \tau) \beta(\tau),
\end{align*}

where $\alpha(\tau)$ and $\beta(\tau)$ are the solutions of the 8-dimensional ODE system

\begin{equation}
\label{eq:odealphabeta}
\begin{split}
\alpha' &=  \omega^{-1} \sin(\omega \tau) \nabla R( \cos(\omega \tau) \alpha+ \omega^{-1} \sin(\omega \tau) \beta), \\
\beta' &= -\cos(\omega \tau)  \nabla R( \cos(\omega \tau) \alpha+ \omega^{-1} \sin(\omega \tau) \beta),
\end{split}
\end{equation}

supplemented with the initial conditions

\begin{equation*}
\alpha(0) = u_0,  \quad
\beta(0) = u'_0.
\end{equation*}



```julia
α0 = u0
β0 = w0
U0 = vcat(α0, β0)
μ = 398600.8
h = μ/norm(q0) - 0.5*dot(v0,v0) -  V(q0)
ω = sqrt(h/2)
C = 1.7554962315534863e10
p = [C, ω, μ];
```

```julia
function J2VOPODE!(dU, U, p, τ)  
    C = p[1]
    ω = p[2]
    θ = ω*τ
    α1 = U[1]
    α2 = U[2]
    α3 = U[3]
    α4 = U[4]
    β1 = U[5]
    β2 = U[6]
    β3 = U[7]
    β4 = U[8]
    c = cos(θ)
    s = sin(θ)/ω
    u1 = c * α1 + s * β1
    u2 = c * α2 + s * β2
    u3 = c * α3 + s * β3
    u3 = c * α3 + s * β3
    u4 = c * α4 + s * β4
    z = 2*(u1*u3 + u2*u4)
    r = u1^2 + u2^2 + u3^2 + u4^2
    w = 1/r^3
    sinth = z/r
    A = 0.5*C*w*(1 - 6*sinth^2)
    B = 1.5*C*w*sinth
    gradU1 = A*u1+B*u3
    gradU2 = A*u2+B*u4
    gradU3 = A*u3+B*u1
    gradU4 = A*u4+B*u2
    dU[1] = s*gradU1
    dU[2] = s*gradU2
    dU[3] = s*gradU3
    dU[4] = s*gradU4
    dU[5] = -c*gradU1
    dU[6] = -c*gradU2
    dU[7] = -c*gradU3
    dU[8] = -c*gradU4
    return nothing
end


function J2VOPEnergy(U,p,τ)
    C = p[1]
    ω = p[2]
    μ = p[3]
    θ = ω*τ
    α1 = U[1]
    α2 = U[2]
    α3 = U[3]
    α4 = U[4]
    β1 = U[5]
    β2 = U[6]
    β3 = U[7]
    β4 = U[8]
    c = cos(θ)
    s = sin(θ)/ω
    s_ = -ω^2 * s 
    u1 = c * α1 + s * β1
    u2 = c * α2 + s * β2
    u3 = c * α3 + s * β3
    u4 = c * α4 + s * β4
    w1 = s_ * α1 + c * β1
    w2 = s_ * α2 + c * β2
    w3 = s_ * α3 + c * β3
    w4 = s_ * α4 + c * β4
    z = 2*(u1*u3 + u2*u4)
    r = u1^2 + u2^2 + u3^2 + u4^2
    ww = w1^2 + w2^2 + w3^2 + w4^2
    u = [u1,u2,u3,u4]
    q = L(u)*u
    return 0.5 * (ww + ω^2 * r) + R(u)
#    return -μ/r + 2*ww/r +  V(q) 
end
```

### Inplementazio bektoriala (VecArray erabiliz) eta Vern9-ren arteko konparaketa

```julia
s=8

t0 = 0.
sasi_periodo = 2*π/ω
sasi_periodo = 6/ω
dt = sasi_periodo/5
tF = 20000*sasi_periodo


n = 1000
#n = 5

m = convert(Int64,ceil(abs(tF-t0)/(n*dt)))
m=100
n = convert(Int64,ceil(abs(tF-t0)/(m*dt))) # Number of macro-steps (Output is saved for n+1 time values)
dt = (tF-t0)/(n*m)
println("dt = $dt, n=$n, m=$m")


prob = ODEProblem(J2VOPODE!, U0, (t0,tF), p)


sol0=IRKGL_Seq(s, U0, t0, tF, n, m, J2VOPODE!, p, initial_interp=false)

#sol1=IRKGL_simd(s, U0, t0, tF, n, m, J2VOPODE!, p, initial_interp=false,
#                 itermax=itermax)

sol1,iters1=solve(prob,IRKGL_simd(s=s, dim=dim, m=m,initial_interp=0, myoutputs=true), dt=dt);

#sol_ex=IRKGL_simd(s, U0, t0, tF, n, 2m, J2VOPODE!, p, initial_interp=false,
#                 itermax=itermax);

sol_ex,iters_ex=solve(prob,IRKGL_simd(s=s, dim=dim, m=2m, initial_interp=0, myoutputs=true), dt=dt/2) ;
```

```julia
yrange = (1e-13,1e-1)
year = 365.5


function energy_plot(sol; label="", legend=:bottom)
    energies = [J2VOPEnergy(sol.u[j],p,sol.t[j]) for j in eachindex(sol.t)]
    E0 = energies[1]
    epsilon = eps(1e-3)
    errors = Float64.(abs.(energies[2:end]/E0 .- 1)) .+ epsilon
    tt = sol.t[2:end]/year
    yrange = (minimum(errors), maximum(errors))
    pl = plot(title="Energiaren errore erlatiboen eboluzioa",
         yscale=:log10, ylims=yrange, legend=legend)
    plot!(tt,  errors, label=label)
    return pl
end


pl0 = energy_plot(sol0, label="IRKGL (sekuentziala)")

pl1 = energy_plot(sol1, label="IRKGL (VecArray)")

pl2 = energy_plot(sol2, label="Vern9")

plot(pl0, pl1,  pl2, layout=(3,1))
```
