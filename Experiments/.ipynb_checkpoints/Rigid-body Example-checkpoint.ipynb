#  Rigid-body problem

```julia
using LinearAlgebra,Plots
using DiffEqDevTools,BenchmarkTools
using OrdinaryDiffEq, Parameters
using JLD2, FileIO
```

```julia
PATH_SRC="../src/"
include(string(PATH_SRC,"MyBenchmarksTools.jl"))
include(string(PATH_SRC,"IRKGL_SIMD.jl"))
using .IRKGL_SIMD
```

## Initial value problem

```julia
function RigidBody!(du,u,p,t)
  I1=p[1]
  I2=p[2]
  I3=p[3]      
  du[1]  = I1*u[2]*u[3]
  du[2]  = I2*u[1]*u[3]
  du[3]  = I3*u[1]*u[2] +  0.25*sin(t)^2
end 

p = [-2.0,1.25,-0.5]
u0 = [1.0;0.0;0.9]

dim=length(size(u0))
prob = ODEProblem(RigidBody!,u0,tspan,p);
```

```julia
t0=0.
tF=100.0 
dt=0.02
tspan=(t0,tF)

s=8
sol = solve(prob,IRKGL_simd(s=s, dim=dim,initial_interp=0),dt=dt);
```

```julia
plot(sol, xlabel="t", title="Rigid Body problem")
```
