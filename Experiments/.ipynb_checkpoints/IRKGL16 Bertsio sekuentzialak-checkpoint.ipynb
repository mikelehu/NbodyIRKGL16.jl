## Outer-solar-system

We consider $N=5$ bodies of the outer solar system: the Sun, Jupiter, Saturn, Uranus, and Neptune.
The initial values $u_{00}$ are taken from DE430, Julian day (TDB) 2440400.5 (June 28, 1969). 

```julia
PATH_ODES="../ODEProblems/"

include(string(PATH_ODES,"Initial5Body.jl"))
include(string(PATH_ODES,"Nbody.jl"))
include(string(PATH_ODES,"Nbody2nd.jl"))
include(string(PATH_ODES,"NbodyDyn.jl"));
```

```julia
u0, Gm, bodylist = Initial5Body(Float64)
u0_B, Gm_B, bodylist = Initial5Body(BigFloat)

q0=u0[:,:,1]
v0=u0[:,:,2]

dim=length(size(u0))
N = length(Gm)

show(bodylist)
E0=NbodyEnergy(u0,Gm)
```

```julia
t0=0.
tF=1e6  # days

tspan= (t0,tF)
tspan_B=(BigFloat(t0),BigFloat(tF))

dim=length(size(u0))
prob = ODEProblem(NbodyODE!, u0,tspan , Gm)
prob2nd = SecondOrderODEProblem(NbodyODE2nd!,v0,q0,tspan,Gm)
probDyn= DynamicalODEProblem(NbodyODEq!,NbodyODEv!,q0,v0,tspan,Gm)
prob_B = ODEProblem(NbodyODE!, u0_B, tspan_B, Gm_B)

#sol= solve(prob,IRKGL16(), dt=200., adaptive=false);
#sol= solve(probDyn,IRKGL16(), dt=200., adaptive=false);
sol = solve(prob,Vern9(),adaptive=false,dt=200.);
```

```julia
pl = plot(title="Outer Solar System", 
            xlabel="x", ylabel="y",  aspect_ratio=1)

for j = 1:5
    x  = [u[1,j,1] for u in sol.u]
    y  = [u[2,j,1] for u in sol.u] 
    pl = plot!(x,y, label=bodylist[j]) 
end 
display(pl)
```

```julia
function energy_plot(sol; title="", label="")
    year=365.25
    yrange=1e-18
    energies = [NbodyEnergy(BigFloat.(u),Gm) for u in sol.u]
    E0 = energies[1]
    epsilon = eps(1e-3)
    errors = Float64.(abs.(energies[2:end]/E0 .- 1)) .+ epsilon
    tt = sol.t[2:end]/year
    pl = plot(title=title,
         yscale=:log10, ylims=yrange)
    plot!(tt,  errors, label=label)
    return pl
end


pl = energy_plot(sol, title="Energy error: 5-body problem", label="Vern9")
plot(pl)
```

### First Order ODE

```julia
tols=abstols=reltols=[1e-8, 1e-9, 1e-10, 1e-11, 1e-12, 1e-13, 1e-14, 1e-15]

dts8= [100., 200., 250., 400.,  500., 625.];

1.e6./dts8
```

```julia
nruns=1000

wp1=launch_IRKGL16_tests(IRKGL16(),final_state, prob, dts8; nruns=nruns)
#
s=8
wp2=launch_IRKGL_seq_tests(final_state, prob, s, dts8; initial_interp=true, nruns=nruns)
#
s=8
wp3=launch_IRKGL_simd_tests(final_state, prob, dim, s, dts8,
                            initial_interp=-1, partitioned=false, nruns=nruns);
```

```julia
plot(title="Outer Solar System benchmark",xlabel="Error", ylabel="Time (s)")

plot!(wp1.errors, wp1.times, seriestype=:scatter, scale=:log10, label="IRKGL16",color="red")
plot!(wp1.errors, wp1.times, scale=:log10, label="", lw=2, color="red")

plot!(wp2.errors, wp2.times, seriestype=:scatter, scale=:log10, label="IRK16_seq",color="blue")
plot!(wp2.errors, wp2.times, scale=:log10, label="", lw=2, color="blue")

plot!(wp3.errors, wp3.times, seriestype=:scatter, scale=:log10, label="IRKGL16_simd",color="green")
plot!(wp3.errors, wp3.times, scale=:log10, label="", lw=2, color="green")
```
