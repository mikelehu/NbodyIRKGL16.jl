```julia
setups = [
          Dict(:alg=>Vern9())
]

solnames =["Vern9"]
wpp1 = WorkPrecisionSet(prob,abstols,reltols,setups;appxsol=test_sol,save_everystep=false,numruns=10,maxiters=10^9, names=solnames)

setups = [
          Dict(:alg=>DPRKN12())   
]
solnames = ["DPRKN12"]
wpp2 = WorkPrecisionSet(prob2nd,abstols,reltols,setups;appxsol=test_solx2nd,save_everystep=false,numruns=10,maxiters=10^9,names=solnames);
```

```julia
plot(title="BenchmarkTools",xlabel="Error", ylabel="Time (s)")
#
plot!(wpp1.wps[1].errors,wpp1.wps[1].times, seriestype=:scatter, scale=:log10, label="Vern9",color="red")
plot!(wpp1.wps[1].errors,wpp1.wps[1].times, scale=:log10,label="", lw=2, color="red")
#
#
plot!(wpp2.wps[1].errors,wpp2.wps[1].times, seriestype=:scatter, scale=:log10, label="DPRKN12",color="green")
plot!(wpp2.wps[1].errors,wpp2.wps[1].times, scale=:log10, label="", lw=2, color="green")
```
